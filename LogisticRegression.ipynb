```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.model_selection import KFold
```

```python
# Load the datasets
df = pd.read_csv('Dataset.csv')

X_train = pd.read_csv("X_train_bin.csv")
X_test = pd.read_csv("X_test_bin.csv")
y_train = pd.read_csv("y_train_bin.csv").squeeze()
y_test = pd.read_csv("y_test_bin.csv").squeeze()
```

```python
from sklearn.model_selection import train_test_split

# Validation Split
X_train2, X_val, y_train2, y_val = train_test_split(X_train, y_train, test_size=0.20, stratify=y_train, random_state=42)
```

```python
best_score = -1
best_params = None
best_model = None
```

```python
# GridSearch 
for C in [0.01,0.1,1,10]:
    for penalty in ["l1", "l2"]:
        gs_model = LogisticRegression(
            C= C,
            penalty= penalty,
            solver= "liblinear",
            max_iter= 5000
        )
        gs_model.fit(X_train2, y_train2)
        gs_pred = gs_model.predict(X_val)
        gs_score = f1_score(y_val, gs_pred, average="weighted")
        if gs_score > best_score:
            best_score = gs_score
            best_params = {"C": C, "penalty": penalty}
            best_model = gs_model
```

```python
print(best_params)
print(best_score)
print(best_model)
```

```python
# GridSearch predictions
gs_pred = best_model.predict(X_test)
```

```python
# calculate evaluation for GridSearch
gs_accuracy = accuracy_score(y_test, gs_pred)
gs_recall = recall_score(y_test, gs_pred)
gs_precision = precision_score(y_test, gs_pred)
gs_f1 = f1_score(y_test, gs_pred)

print(f"GrisSearch Accuracy: {gs_accuracy}")
print(f"GrisSearch Recall: {gs_recall}")
print(f"GrisSearch Precision: {gs_precision}")
print(f"GrisSearch F1 Score: {gs_f1}")
```

```python
model = LogisticRegression(max_iter= 5000)
param_grid = [
    {"penalty" : ["l1", "l2"],
     "C" : [0.01,0.1,1,10],
     "solver": ["liblinear"]}
]
```

```python
from sklearn.model_selection import GridSearchCV

clf = GridSearchCV(model, param_grid = param_grid, cv= 3, scoring= "f1_weighted", verbose= True, n_jobs= 1)
```

```python
best_clf = clf.fit(X_train, y_train)
```

```python
cv_pred = best_clf.best_estimator_.predict(X_test)
```

```python
# calculate evaluation for GridSearchCV
cv_accuracy = accuracy_score(y_test, cv_pred)
cv_recall = recall_score(y_test, cv_pred)
cv_precision = precision_score(y_test, cv_pred)
cv_f1 = f1_score(y_test, cv_pred)

print(f"GridSearchCV Accuracy: {cv_accuracy}")
print(f"GridSearchCV Recall: {cv_recall}")
print(f"GridSearchCV Precision: {cv_precision}")
print(f"GridSearchCV F1 Score: {cv_f1}")
```
