```python
import pandas as pd
import glob
import kagglehub
import numpy as np
```

```python
# Download latest version
path = kagglehub.dataset_download("chethuhn/network-intrusion-dataset")

print("Path to dataset files:", path)
```

```python
# Load the data
all_df = []
for file_name in glob.glob(path + "/*.csv"):
    print(f'Loading {file_name}')
    new_df = pd.read_csv(file_name)
    all_df.append(new_df)
print(f"Files Loaded: {len(all_df)}")

# Concat all the data
df = pd.concat(all_df)
```

```python
# Remove spaces in start and end, all column names in lower case, empty spaces with "_"
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
# # Show all the columns & rows
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

# Filling NaN with the mean
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df = df.fillna(df.mean(numeric_only=True))

# Checking NaN in every column
print(df.isnull().sum())
```
